```python
import numpy as np

scores=np.random.randint(50,101,size=(50,5))
scores
```

```python
import pandas as pd

df=pd.DataFrame(scores,
                index=[f"學生{i+1}" for i in range(50)],
                columns=["國文","英文","數學","地理","歷史"])
df
```

```python
#series(1維)，dataframe(2維)
#透過欄的選取

df["國文"]
df["英文"]
df[["國文","英文"]]
```

```python
#透過index的名稱選取
df.loc["學生1"]
df.loc[["學生1","學生2"]]
```

```python
#透過index的索引值選取
df.iloc[0]
df.iloc[[0,1]]
```

```python
df.iloc[5:]
df.iloc[-5:]
```

```python
import numpy as np
total=np.sum(df,axis=1)
mean=np.mean(df,axis=1)
```

```python
df["總分"]=total
df["平均"]=mean
```

```python
five_df=df.iloc[:5]
five_df[five_df["總分"]>=380]
```
