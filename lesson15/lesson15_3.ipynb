```python
import numpy as np
one_store=np.random.randint(100,10000,size=(30,))
one_store
```

```python
#累計每天這家店的營收
oneStoreCum=one_store.cumsum()
oneStoreCum
```

```python
import matplotlib.pyplot as plt
figure=plt.figure(figsize=(10,5))
axes=figure.add_subplot(1,1,1)
axes.plot(range(1,31),oneStoreCum,marker='o',linestyle='-',color='blue',label='Cumulative Revenue')
axes.set_title('Cumulative Revenue of One Store')
axes.set_xlabel('Day')
axes.set_ylabel('Revenue')
axes.legend()
axes.grid()
plt.show()
```
