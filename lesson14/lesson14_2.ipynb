```python
#載入字型
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont('ChineseFont.ttf')  # 字型檔路徑
mpl.rc('font', family='ChineseFont')
```

```python
import matplotlib.pyplot as plt
# 來源資料
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())
print(scores1, scores2)

# 建立圖形和軸
figure = plt.figure(figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k')
axes = figure.add_subplot(111)
axes.plot(subjects, scores1, label='學生1', marker='o', color='blue')
axes.plot(subjects, scores2, label='學生2', marker='o', color='red')
axes.set_title('學生成績比較', fontsize=16)
axes.set_xlabel('科目', fontsize=14)
axes.set_ylabel('分數', fontsize=14)
axes.grid(True, linestyle='--', alpha=0.5,axis='y')
#legend,default
axes.legend(loc='best', fontsize=10,edgecolor='black', facecolor='lightgray')
#marker上顯示學生分數
for i, score in enumerate(scores1):
    axes.text(subjects[i], score+1, str(score), ha='center', va='bottom', fontsize=10, color='blue')
for i, score in enumerate(scores2):
    axes.text(subjects[i], score + 1, str(score), ha='center', va='bottom', fontsize=10, color='red')
# marker上顯示學生分數,超過圖形範圍,將圖形範圍調整
# 設定y軸範圍(使用原有的最大值和最小值,各增加5)
axes.set_ylim( min(min(scores1), min(scores2))- 5, max(max(scores1), max(scores2)) + 10)
plt.tight_layout()  # 自動調整子圖參數
plt.show()
```
